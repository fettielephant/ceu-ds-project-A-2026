```python
import pandas as pd

#loading the data
df = pd.read_csv("Sanctions_Russia_Cleaned_Final.csv")
#date column
df["Listed On"] = pd.to_datetime(df["Listed On"], errors="coerce")
# year and month period
df["month"] = df["Listed On"].dt.to_period("M")

#aggregate uinique people monthly
monthly_unique_people = (df.groupby("month")["Group ID"].nunique().reset_index(name="unique_sanctioned_people"))
monthly_unique_people["month"] = monthly_unique_people["month"].dt.to_timestamp()
monthly_unique_people = monthly_unique_people.sort_values("month")

#csv create
monthly_unique_people.to_csv("sanctioned_people_by_month.csv",index=False)

```
