```python
import pandas as pd
import numpy as np
```

```python
ftse_100 = pd.read_csv('ftse_100.csv')
ftse_350 = pd.read_csv('ftse_350.csv')
ftse_aim_100 = pd.read_csv('ftse_aim_100.csv')
ftse_aim_AS = pd.read_csv('ftse_aim_AS.csv')
ftse_aim_UK_50 = pd.read_csv('ftse_aim_UK_50.csv')
ftse_all_share = pd.read_csv('ftse_all_share.csv')
ftse_fledging = pd.read_csv('ftse_fledging.csv')
ftse_midcap_250 = pd.read_csv('ftse_midcap_250.csv')
ftse_small_cap = pd.read_csv('ftse_small_cap.csv')
ftse_all_small = pd.read_csv('ftse_all_small.csv')
```

```python
def convert_data(df1):
    '''
    converts the dataframe and does feature engineering
    Input:
    df1 - original dataframe
    Returns:
    df - modified dataframe
    '''
    df_full = df1.copy()
    # rearrange the dataset
    df_full['Date'] = pd.to_datetime(df_full['Date'])
    df = df_full.loc[(df_full['Date'] >= '2013-01-01')]
    df = df.sort_values('Date')

    # strip the data
    df['Price'] = (df['Price'].str.replace(',', '')).astype(float)
    df['Open'] = (df['Open'].str.replace(',', '')).astype(float)
    df['High'] = (df['High'].str.replace(',', '')).astype(float)
    df['Low'] = (df['Low'].str.replace(',', '')).astype(float)
    df['Change in decimals'] = (df['Change %'].str.replace('%', '')).astype(float) / 100
    df.drop('Change %', axis = 1, inplace = True)
    if df['Vol.'].isna().sum() > 0:
        df['Vol.'] = df['Vol.'].fillna(0)
    else:
        df['Vol.'] = (df['Vol.'].str.replace('K', '0').str.replace('M', '0000').str.replace('B', '0000000').str.replace('.', '')).astype(int)

    return df
```

```python
data = [ftse_100, ftse_350, ftse_aim_100, ftse_aim_AS, ftse_aim_UK_50, ftse_all_share, ftse_fledging, ftse_midcap_250, ftse_small_cap, ftse_all_small]
names = ['ftse_100', 'ftse_350', 'ftse_aim_100', 'ftse_aim_AS', 'ftse_aim_UK_50', 'ftse_all_share', 'ftse_fledging', 'ftse_midcap_250', 'ftse_small_cap', 'ftse_all_small']
i = 0
for table in data:
    clean_data = convert_data(table)
    name = names[i] + "_clean.csv"
    clean_data.to_csv(name, index = False)
    i += 1
```
